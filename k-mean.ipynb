

### import Neseccary packages



```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.datasets  import make_blobs
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```



# Step 1: Data Generation



```python
X,y=make_blobs(n_samples=2000,random_state=130)
```

```python
X.shape
```

```python
k = 5
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_ 
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c=np.unique(y_pred),
            edgecolors='k', linewidths=0.5, label='Centroids')
plt.title(f'K-Means Clustering with k= 5', fontsize=14)
plt.xlabel('Feature 1', fontsize=12)   # Set x-axis label
plt.ylabel('Feature 2', fontsize=12)   # Set y-axis label
kmeans.inertia_
```


# Elbow Method for Optimal Number of Clusters (k)


```python
wcss_list= []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++',n_init='auto', random_state= 42)
    kmeans.fit(X)
    wcss_list.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss_list)
plt.title('The Elobw Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('wcss_list')
plt.show();
```

# K-Means Clustering with Selected k

```python
k = 3
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_ 
```

```python
plt.figure(figsize=(10, 6)) # Adjust figure size
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', label='Data Points')  # Plot datapoints with predicted labels
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c=np.unique(y_pred),
            edgecolors='k', linewidths=2, label='Centroids') # Plot centroids
plt.title(f'K-Means Clustering with k={k}', fontsize=14)  # Set title
plt.xlabel('Feature 1', fontsize=12)   # Set x-axis label
plt.ylabel('Feature 2', fontsize=12)   # Set y-axis label
plt.grid(True, linestyle='--')  # Add grid to improve readability
plt.legend()   # Show legend
plt.tight_layout()  # Adjust spacing
plt.show() # Show the k-means plot.
```
